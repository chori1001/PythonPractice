```python
for item in names:
    tem = str(item)
    print(tem)
```

```python
game_name_list = []

for item in names:
    name = item.text.split('.')
    if(len(name) < 3) :        
        game_name = name[0]
        game_name_list.append(game_name)
            
result = pd.DataFrame(data={'Apple_GameName' : game_name_list})
```
